1. (5p)Escriba un programa que permita generar N = 10000 números
aleatorios distribuidos de acuerdo a la densidad de probabilidades

$p(x)={Ae^{-x}}$   si $0<= x < 1$ y 0 en otros casos 

mediante el método de transformación. $A$ es una constante de normalización.

Mostrar las distribuciones de $r$ y $x$ en forma de histogramas.

DESCOMPONIENDO SE HALLA $A$

$A = 1/(1-exp(-1))$ o $A=\frac{e}{e-1}$
Trabajo con la primera


```python
# -*- coding: utf-8 -*-
"""

Obtener la distribucion de NA de acuerdo a p(x)=exp(-x)
mediante el metodo de inversion
x = -ln(1-r)
@author: justo
"""

import numpy as np
from random import random
import matplotlib.pyplot as plt

A = 1/(1-np.exp(-1))  # una constante de normalización

def Transformacion(N):
    vx=[]
    vr = []
    for i in range(N):
        r = random()  # na entre 0 y 1
        x = -np.log(1-r/A)  # transformacion
        vx.append(x)
        vr.append(r)
    return (vx,vr) 

# numero de eventos (puntos)
N = 10000

# llamado de la funcion
gx,gr =Transformacion(N)



plt.figure(1)
plt.title('distribucion homogenea')
plt.hist(gr,color="red")
plt.xlabel('r')
    

plt.figure(2)
plt.title('Distribucion exponencial')
plt.hist(gx,color="red")
plt.xlabel('x')
```
